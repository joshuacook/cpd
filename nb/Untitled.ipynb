```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.io as sio
test = sio.loadmat('../src/project1/mydata.mat')

results = pd.DataFrame(test["mydata"]).T
results.columns = ["x", "y", "theta", "v", "omega", "ready", "theta_dif", "state"]
n = results.shape[0]
results["time"] = np.linspace(1, n*0.033, n)
min_v, max_v = 0, 100
# min_v, max_v = 40,55
fig, ax = plt.subplots(1, 1, figsize=(20,5))
window = results[(results.time > min_v) & (results.time < max_v)]
window.plot(kind="scatter", y="y", x="time", ax=ax, c="cyan", label="y")
window.plot(kind="scatter", y="x", x="time", ax=ax, c="blue", label="x")
window.plot(kind="scatter", y="theta", x="time", ax=ax, c="red", label="theta")
window.plot(kind="scatter", y="omega", x="time", ax=ax, label="omega")
window.plot(kind="scatter", y="theta_dif", x="time", ax=ax, label="theta_dif", c="green")
window.plot(kind="scatter", y="state", x="time", ax=ax, label="state", c="orange")
ax.legend()

```
